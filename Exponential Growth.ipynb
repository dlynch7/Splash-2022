```python
import numpy as np
from matplotlib import pyplot as plt

### 1. Set up problem ###
def exp_growth(x,c):
    # rate of change of x with respect to t is proportional to current value of x:
    dx_dt = c*x
    return dx_dt

x0 = 1 # initial condition
c = 1  # growth coefficient
# TODO: experiment with different values for N (11,101,1001,...).
# How does Npts affect the accuracy of the numerical solution?
N = 11
tspan = np.linspace(0,10,N) # a list of times at which to compute x, given the model defined above
dt = tspan[1] - tspan[0] # timestep size

x = [x0] # we will use this list to store the predicted value of x at each time in tspan
x_sol = [x0] # we will use this list to store the actual value of x at each time in tspan



### 2. Simulate (integrate) forward in time ###
for t in tspan[0:-1]:
    dx_dt = exp_growth(x[-1],c) # compute the rate of change of x with respect to time
    x_next = x[-1] + dx_dt*dt # approximate the next value of x, dt into the future
    x.append(x_next) # store the next value of x in a list for plotting
    
    x_sol.append(x0*np.exp(c*t))
    

    
### Visualize results: ###
plt.rcParams['figure.figsize'] = [20,10]
plt.rcParams["font.size"] = "18"

plt.plot(tspan,x,'b-',label='numerical solution')
plt.plot(tspan,x_sol,'r-',label='exact (analytical) solution')
plt.xlabel('t')
plt.ylabel('x')
plt.legend(loc='best')
plt.grid()
plt.show()
```
